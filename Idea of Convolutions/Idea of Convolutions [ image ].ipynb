```python
import cv2

img = cv2.imread('sample.jpg')

from matplotlib import pyplot as plt

plt.imshow(img)
```

```python
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
#gray -> result image, img -> source image
#cv2.COLOR_BGR2GRAY -> color conversion method
#cvtColor function is used to convert color

plt.imshow(gray)
```

```python
plt.imshow(gray,cmap='gray')
```

```python
import numpy as np

kernel = np.array([[0.75,0,0.75],
                 [0.5,0,0.5],
                 [0.75,0,0.75]],np.float)

conv = cv2.filter2D(gray,-1,kernel)

plt.imshow(conv,cmap='gray')
```

```python
kernel_2 = np.array([[1, 2, 1], [0, 0, 0], [-1, -2, -1]], np.float)
conv_2 = cv2.filter2D(gray, -1, kernel_2)
plt.imshow(conv_2, cmap = 'gray')
```

```python
kernel_3 = np.array([[1, 0, -1], [2, 0, -2], [1, 0, -1]], np.float)
conv_3 = cv2.filter2D(gray, -1, kernel_3)
plt.imshow(conv_3, cmap = 'gray')
```

```python
kernel_4 = np.array([[1, 2, 1], [1, 2, 0], [-1, -2, -1]], np.float)
conv_4 = cv2.filter2D(gray, -1, kernel_4)
plt.imshow(conv_4, cmap = 'gray')
```
